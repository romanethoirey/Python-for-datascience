```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math

from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("./feature_engineering1.csv")
```

```python
model = LinearRegression()
xp = df[['experience']]
sal = df[['salary']]
```

```python
model.fit(xp, sal)
prediction_salary = model.predict(xp)
plt.scatter(xp, prediction_salary, c='red', marker='+')
plt.scatter(xp, sal, c='blue', marker='.')
plt.show()
```

```python
salcos = np.cos(sal)
model.fit(xp, salcos)
prediction_salary = model.predict(xp)
plt.scatter(xp, prediction_salary, c='red', marker='+')
plt.scatter(xp, salcos, c='blue', marker='.')
plt.show()
```

```python
salsqrt = np.sqrt(sal)
model.fit(xp, salsqrt)
prediction_salary = model.predict(xp)
plt.scatter(xp, prediction_salary, c='red', marker='+')
plt.scatter(xp, salsqrt, c='blue', marker='.')
plt.show()
# best transformation
```

```python
sallog = np.log(sal)
model.fit(xp, sallog)
prediction_salary = model.predict(xp)
plt.scatter(xp, prediction_salary, c='red', marker='+')
plt.scatter(xp, sallog, c='blue', marker='.')
plt.show()
```

```python
salexp = np.exp(sal)
model.fit(xp, salexp)
prediction_salary = model.predict(xp)
plt.scatter(xp, prediction_salary, c='red', marker='+')
plt.scatter(xp, salexp, c='blue', marker='.')
plt.show()
# marche po
```

**On training models**:  
When you do a fit on a linear regression, le paramètre $\theta$ est trouvé automatiquement.  
GD to minimise the loss function (MSE).

Problem:  
if the features do not have the same scale. GD will not work well.

<font color='green'>ex</font>: predict if a land ill be sold in 6 months  
Dataset = size (big number) type (one hot encoding)  
Size is much bigger than Type and it will prevent gradient descent working well  -> Scale Down Numerical values

Ways to normalize variable:  
* standard scaler X -> X - mean(X)  
Most values (but not all) will be close to 0
* Min Max Scaler X -> $\frac{X - min(X)}{max(X)-min(X)}$  
all values will be between 0 and 1

<font color='red'> Remember to scale feature. For gradient based algorithm.</font>
