# Clustering practical


Load the 'dataset_1.csv'

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import SpectralClustering
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
dataset_1 = pd.read_csv('./dataset_1.csv', index_col = 0)
```

Display the scatter plot of the data

```python
plt.scatter(dataset_1['normalized_age'], dataset_1['normalized_salary'])
```

Look on the documentation on how to run Kmeans and run a kmeans model with k = 2 on the dataset

```python
Kmean = KMeans(n_clusters=2)

Kmean.fit(dataset_1)
Y = Kmean.predict(dataset_1)

Kmean.cluster_centers_
```

```python
Spectral_clustering = SpectralClustering(n_clusters=2)
Spectral_clustering.fit(dataset_1)

Y = Spectral_clustering.fit_predict(dataset_1)

plt.scatter(dataset_1['normalized_age'], dataset_1['normalized_salary'], c = Y)
plt.show()
```

Display each point of the data with a color depending on the cluster

```python
plt.scatter(dataset_1['normalized_age'], dataset_1['normalized_salary'], c = Y)
plt.show()
```

## A second dataset

Load the the dataset_2.csv and display the scatter plot

```python
dataset_2 = pd.read_csv('./dataset_2.csv', index_col=0)
plt.scatter(dataset_2['normalized_age'], dataset_2['normalized_salary'])
```

With a for loop, train a kmeans model with k = 1, 2, ..., 20 and each time display the scatter plot and the loss of the model

```python
scaler = StandardScaler()
dataset_2 = scaler.fit_transform(dataset_2)
dataset_2 = pd.DataFrame(dataset_2, columns=['normalized_age', 'normalized_salary'])

scores = []

for i in range(1,20):
    Kmean = KMeans(n_clusters=i)
    Kmean.fit(dataset_2)
    
    predict = Kmean.predict(dataset_2)
    
    plt.scatter(dataset_2['normalized_age'], dataset_2['normalized_salary'], c = predict)
    plt.show()
    
    scores.append(abs(Kmean.score(dataset_2)))
    
    print('The loss of the', i, 'mean model is :', Kmean.score(dataset_2))
```

```python
plt.plot(scores)
```

How does the loss of the model change when you increase k ? 

It diminues.

```python
dataset_3 = pd.read_csv('./dataset_3.csv', index_col=0)

Kmean = KMeans(n_clusters=2)
Kmean.fit(dataset_3)

Y = Kmean.predict(dataset_3)
plt.scatter(dataset_3['x1'], dataset_3['x2'], c = Y)
plt.show()
```

Run the Spectral clustering of scikit learn on it and display the results

```python
Spectral_clustering = SpectralClustering(n_clusters=2, eigen_solver='arpack', affinity='nearest_neighbors')
Spectral_clustering.fit(dataset_3)

Y = Spectral_clustering.fit_predict(dataset_3)

plt.scatter(dataset_3['x1'], dataset_3['x2'], c = Y)
plt.show()
```

Run the dbscan algorithm on it and display the results

```python
Dbscan = DBSCAN(eps=0.3, min_samples=2)

Dbscan.fit(dataset_3)
Y = Dbscan.fit_predict(dataset_3)

plt.scatter(dataset_3['x1'], dataset_3['x2'], c = Y)
plt.show()
```
