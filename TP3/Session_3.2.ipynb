## Time series data

Pandas was originaly developed for time series data and is very good for them. 



Generate a dataframe containing a datetime column starting from January the first 2019 until March the first 2019 with each row representing an hour. The Datetime should be the index of the dataframe. In the following you may be interested in the special attribute .dt of timeseries dataframe. What does it do ? What does it contain ? 

Generate a second column containing random values for each row of the time serie

```python
hours = pd.date_range('01-01-2019', '03-01-2019', freq='H')
datetime = pd.DataFrame({'Val': np.random.randn(len(hours))})
datetime = datetime.set_index(hours, drop=True, append=False, inplace=False, verify_integrity=False)
datetime
```

Compute the mean for each day of the month

```python
datetime.groupby(datetime.index.day).mean() # methode resample pour les jours du mois c'est bien avec les times sÃ©ries
```

Compute the median for each day of the week

```python
datetime.groupby(datetime.index.week).median()
```

Select all values from february

```python
datetime[datetime.index.month == 2]
```

Compute the moving average of the random values with a window of three hours. SÃ©rie of means on a small portion of the data.

```python
datetime.rolling(window=3).mean()
```

## Weather data

Load the following data set which contains time series data of the weather in Toronto. When you do read_csv you can specify to parse date for several columns. You can also decide to make a given column as an index. Parse the date and put as the index

```python
weather = pd.read_csv('data_weather_montreal.csv', index_col ='Date/Time', parse_dates=True)
weather
```

Display the columns

```python
weather.columns
```

Rename the column Temp (Â° C) into temperature

```python
weather.rename({'Temp (Â°C)': 'temperature'}, inplace=True, axis='columns')
```

Plot the Temperature versus the time

```python
import matplotlib.pyplot as plt

plt.plot(weather.index, weather['temperature'])
```

Compute the mean and median temperature for each hour of the day and plot it

```python
daily_temp = weather['temperature'].resample('d', how='mean')
plt.plot(daily_temp)
```
