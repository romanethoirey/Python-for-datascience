## join 


Like in SQL, you can join two dataframes. Load the dataset orders.csv and transactions.csv in two different dataframe. The datasets represents orders and order_items. 

```python
import pandas as pd
```

```python
items = pd.read_csv('items.csv', index_col="id") #file transaction renamed to items
```

```python
items.head()
items.sort_values(by='order_id')
```

Explore quickly the data

```python
orders = pd.read_csv('orders.csv', index_col="id")
orders.head()
```

```python
orders.sort_values(by='customer_id')
```

```python
import numpy as np
order_totals = items.groupby('order_id').agg(np.sum)

# or items.groupby('order_id').sum()

order_totals.head()
```

Compute the average order amount => it should output one number.

```python
order_totals.mean()
```

Compute the averate order amount (total) per customer

```python
joined = items.join(orders, on='order_id')
joined.head()
```

```python
per_customer = joined.groupby(['customer_id', 'order_id']).sum()
```

```python
per_customer.head()
```

```python
(1092 + 863) /2
```

```python
per_customer.head().groupby('customer_id').mean()
```
