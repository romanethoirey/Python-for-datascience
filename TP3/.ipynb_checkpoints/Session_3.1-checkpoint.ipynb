# Pandas

```python
import pandas as pd
import numpy as np
```

## Join 


Like in SQL, you can join two dataframes. Load the dataset orders.csv and transactions.csv in two different dataframe. The datasets represents orders and order_items. 

```python
orders = pd.read_csv('orders.csv')
transactions = pd.read_csv('transactions.csv')
items = pd.read_csv('items.csv')
```

Explore quickly the data

```python
orders # id, customer_id
transactions # id, order_id, price
```

Each order has several transactions (or better put) several items. Compute the total amount of each order. One row = one order with the total of all items

```python
transactions.groupby('order_id').agg(np.sum)[['price']]
```

Compute the average order amount => it should output one number.

```python
np.mean(transactions['price'])
```

Compute the average order amount (total) per customer

```python
total = pd.merge(transactions, orders)
total.groupby('customer_id').agg(np.sum)[['price']]
```
