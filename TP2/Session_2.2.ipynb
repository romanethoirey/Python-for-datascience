## Masks revisions

With the randn method of numpy create a vector with 100 values

```python
v = np.random.randn(100)
```

With a mask, select all values above 0.1 and store it in a vector v2

```python
v2 = v[ v > 0.1 ]
```

with a mask select all values above 0.1 or below -0.1 and store it in a vector v3

```python
v3 = v[(v > 0.1) | (v < -0.1) ]
```
