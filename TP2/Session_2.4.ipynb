# A bit of machine learning 



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

What are train set and validation sets ? 

<font color='green'>Réponse : </font> A validation dataset is a sample of data held back from training your model that is used to give an estimate of model skill while tuning model’s hyperparameters.

The validation dataset is different from the test dataset that is also held back from the training of the model, but is instead used to give an unbiased estimate of the skill of the final tuned model when comparing or selecting between final models.

Do a train test split with scikit learn on the dataset you cleaned

```python
df = pd.read_csv('./train.csv')
X_train, X_test = train_test_split(df, test_size=0.33)
```

Train a logistic regression to predict Survived with the variable Age, Sex and the One hot encoded Pclasses

```python
df['Sex'].replace(to_replace=['female','femme'],value=1, inplace=True)
df['Sex'].replace(['male', 'homme'], 0 , inplace=True)
df['Survived'].replace(['O'], 0, inplace=True)
df['Survived'].fillna(0, inplace=True)
df['Survived']= df['Survived'].astype('int')
```

```python
df = pd.concat([df, pd.get_dummies(df['Pclass'], prefix='Pclass'), pd.get_dummies(df['Sex'], prefix='Sex')], axis=1)
df.fillna(0, inplace=True)
df.drop(columns='Pclass')
```

```python
model = LogisticRegression()
X = df[['Age', 'Sex_0','Sex_1', 'Pclass_1', 'Pclass_2', 'Pclass_3']]
Y = df['Survived']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3)
model.fit(X_train,Y_train)
model.predict(X_train)
```

Display the accuracy on the train set and the test set. What can you say ? 

```python
print(model.score(X_train, Y_train),'regression train.\n')
print(model.score(X_test,Y_test),'regression test.\n')
```

Load the Decistion tree classifier and do the same. Are the performance identicals ? 

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, Y_train)
prediction_survived_tree = tree.predict(X_train)
```

```python
print(tree.score(X_train, Y_train),'tree train.\n')
print(tree.score(X_test,Y_test),'tree test.\n')
```
