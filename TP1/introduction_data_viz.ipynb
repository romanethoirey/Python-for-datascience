# Datavisualisation with matplot lib

The matplotlib library is usually imported with the following statement

```python
import matplotlib.pyplot as plt
import numpy as np
```

Create a list containing the following values [0.5, 1, 0.5, 2, 3, 5, 6] and store it in the variable **experience**

```python
experience = [0.5, 1, 0.5, 2, 3, 5, 6]
```

Create a list containing the following values [1050, 1500, 1100, 2500, 3000, 4500, 6000]  and store it in the variable **salary** 

```python
salary = [1050, 1500, 1100, 2500, 3000, 4500, 6000]
```

with the method plot of plt plot the salary versus the experience. What happened ? If nothing is show. use the statemont plt.show() after the plot statement

```python
plt.plot(experience, salary)
```

To fix that, it is better to use a scatter plot. Plot the data with a scatter plot 

```python
plt.scatter(experience, salary)
```

create the function f(x) = x^2 and plot the f(x) with matplotlib for x between -5 and 5. Plot it in blue.

```python
def f(x):
    return x*x

x = np.arange(-5,6,1)
plt.plot(x, f(x), c='blue')
```
