With a mask display all the houses whose size is above 80 

```python
above_80 = (df["size"].values > 80)
df[above_80]
```

With a mask display all the houses whose size are above 100 and have more the 2 rooms

```python
above_100_more_2 = (df["size"] > 100) & (df["nb_rooms"] > 2)
df[above_100_more_2]
```
