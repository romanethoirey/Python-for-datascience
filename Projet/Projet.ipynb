```python
import pandas as pd
import numpy as np
import ast

df_keywords = pd.read_csv("the-movies-dataset/keywords.csv")
df_movies_metadata = pd.read_csv("the-movies-dataset/movies_metadata.csv")
```

```python
df_keywords = df_keywords[df_keywords.astype(str)['keywords'] != '[]']
df_keywords = df_keywords.set_index('id')
df_keywords
```

```python
df_genres = pd.concat([df_movies_metadata['id'],df_movies_metadata['genres']], axis = 1)
df_genres = df_genres[df_genres.astype(str)['genres'] != "[]"]
df_genres = df_genres.set_index('id')
df_genres
```

```python
df_keywords.index = df_keywords.index.map(str)
df_genres.index = df_genres.index.map(str)
df_test = pd.merge(df_keywords, df_genres, left_index=True, right_index=True)
```

```python
df_test.index = df_test.index.map(int)
df_test.sort_index()
df_test
```

```python
genres = ['Animation', 'Comedy', 'Family', 'Adventure', 'Fantasy', 'Romance', 'Drama', 'Action', 'Crime', 'Thriller', 'Horror', 'History', 'Science Fiction', 'Mystery', 'War', 'Foreign', 'Music', 'Documentary', 'Western', 'TV Movie', 'Carousel Productions', 'Vision View Entertainment', 'Telescene Film Group Productions', 'Aniplex', 'GoHands', 'BROSTA TV', 'Mardock Scramble Production Committee', 'Sentai Filmworks']

list_keywords = []
list_appearance_genres = []
list_appearance = []

for i in range(len(df_test)):#parcourir le dataframe -> len(df_test)
    list_genres = [0] * 28
    
    line_genres = ast.literal_eval(df_test.iloc[i]['genres'])#convert la ligne en dict
    long_line_genres = len(line_genres)
    
    for j in range(28):#parcourir les genres
        
        for k in range(long_line_genres):#voir s'ils correspondent au film
        
            if str(line_genres[k]['name']) == genres[j]:#si oui, on passe à 1
                list_genres[j] = 1
    
    line_keywords = ast.literal_eval(df_test.iloc[i]['keywords'])#convert la ligne en dict
    long_line_keywords = len(line_keywords)
    
    for j in range(long_line_keywords):#parcourir les keywords
        
        if str(line_keywords[j]['name']) in list_keywords:#voir s'il est déjà dans la lise des keywords
            
            ind = list_keywords.index(line_keywords[j]['name'])
            list_appearance[ind] += 1#incrementer son nombre d'appartion total
            
            for k in range(28):#incrementer son nombre d'apparition par genre
                
                list_appearance_genres[ind][k] += list_genres[k]
            
        else:
            
            list_keywords.append(str(line_keywords[j]['name']))
            list_appearance_genres.append(list(list_genres))
            list_appearance.append(1)
```

```python
df_final = pd.DataFrame(list(zip(list_keywords, list_appearance_genres, list_appearance)), columns =['Keyword', 'AppearancePerGenre', 'AppearanceTotal'])
```
